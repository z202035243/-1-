```python
import numpy as np
```

```python
input_arr = np.array(df_data)
target_arr = np.array(df_labels)
```

```python
print(input_arr.shape)
```

```python
np.random.seed(456)
index = np.arange(569)
np.random.shuffle(index)
```

```python
print(index)
```

```python
X_train.shape_input = input_arr[index[:456]]
Y_train.shape_target = target_arr[index[:456]]
```

```python
X_test.shape_input = input_arr[index[456:]]
Y_test.shape_target = target_arr[index[456:]]
```

```python
print("X_train.shape :", X_train.shape_input)

print("Y_train.shape :", Y_train.shape_target)

print("X_test.shape:", X_test.shape_input)

print("Y_test.shape:", Y_test.shape_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train.shape_input , Y_train.shape_target.ravel())
```

```python
knn.score(X_test.shape_input , Y_test.shape_target)
```
